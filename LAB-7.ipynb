Lab 7 Implement t-distributed Stochastic Neighbor Embedding (t-SNE)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

data = load_digits()
X, y = data.data, data.target


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)


plt.figure()
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.title("t-SNE Visualization (2D)")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.colorbar()
plt.show()


X_train, X_test, y_train, y_test = train_test_split(
    X_tsne, y, test_size=0.2, random_state=42
)

plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], label='Train', alpha=0.6)
plt.scatter(X_test[:, 0], X_test[:, 1], label='Test', alpha=0.6)
plt.title("t-SNE Train vs Test Distribution")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.legend()
plt.show()
```
